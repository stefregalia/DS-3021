## What is the relationship between age and outcome?

### Pseudocode: 
- We will break the dataset into four age groups: Child (1-18), Adult (19-40), Middle Age (41-59), and Senior (60+)
- We will then count the total different outcomes for each of the four age groups and dividing it by the length of that column to get fair proportions.
- Draw conclusions

```python
child = covid.query("age < 19")  # Filtering for people under 19
adult = covid.query("19 <= age <= 40")  # Filtering for people between 19 and 40
middle_age = covid.query("41 <= age <= 59")  # Filtering for people between 41 and 59
senior = covid.query("age >= 60")  # Filtering for of people 60 or older
```

```python
child_outcomes = child.groupby('outcome').size()/len(child) # Calculates the proportion of each outcome for children
adult_outcomes = adult.groupby('outcome').size()/len(adult) # Calculates the proportion of each outcome for adults
middle_age_outcomes = middle_age.groupby('outcome').size()/len(middle_age) # Calculates the proportion of each outcome for middle aged people
senior_outcomes = senior.groupby('outcome').size()/len(senior) # Calculates the proportion of each outcome for seniors
```

```python
# Prints the proportion of discharged and deceased people per age group

print(child_outcomes)
print(adult_outcomes)
print(middle_age_outcomes)
print(senior_outcomes)
```

```python
def age_on_outcome(x):
    child = covid.query("age < 19")  
    adult = covid.query("19 <= age <= 40")  
    middle_age = covid.query("41 <= age <= 59")  
    senior = covid.query("age >= 60")
    child_outcomes = child.groupby('outcome').size()/len(child) 
    adult_outcomes = adult.groupby('outcome').size()/len(adult) 
    middle_age_outcomes = middle_age.groupby('outcome').size()/len(middle_age) 
    senior_outcomes = senior.groupby('outcome').size()/len(senior)

    return child_outcomes, adult_outcomes, middle_age_outcomes, senior_outcomes
    
```
